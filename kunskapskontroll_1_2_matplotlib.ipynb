### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
MATLAB-stilgränssnitt: Detta gränssnitt är tillståndsbaserat och mindre flexibelt. Det är lämpligt för enkla diagram och snabba visualiseringar. Du använder funktioner som plt.plot() och plt.xlabel() för att ändra den aktuella figurens tillstånd.
Objektorienterat gränssnitt: Denna metod är kraftfullare och mer flexibel. Den är idealisk för komplexa diagram och anpassning. Du skapar en figur (fig) och axlar (ax) för att styra egenskaper och plotta data
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
#Figure: The whole picture, it represents the entire figure and can be seen as canvas.

#Axes: The area of the picture where the graph or drawing is.

#Axis: This decides the size and has marks and names on those marks.

#Artist: This is every visible part on the picture, including the graph and marks. Most of these belong to the Axes.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#y = x^2 [-4 < 4]
#b = np.linsace(y = x^2 [-4 < 4])
x = np.linspace(-4, 4, num= 100)
y = x ** 2
plt.plot(x, y)

plt.show()
print(y)
print()
print(x)
```

```python
x = np.linspace(-4, 4,)
y = x ** 2
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
print()
```
